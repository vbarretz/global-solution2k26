# bibliotecas

```python
import pandas as pd
```

# carregando csv

```python
df_meteo = pd.read_csv('C:\Arquivos Vsco\Projeto\Projeto GS 2K26\data\dados_clima_tratados.csv')
df_ndwi = pd.read_csv(r'C:\Arquivos Vsco\Projeto\Projeto GS 2K26\data\bruto\dados_ndwi.csv')


print(f"\n\t Dados climáticos: {len(df_meteo)} registros")
print(f"\n\t Dados NDWI: {len(df_ndwi)} registros\n")
```

# fazendo o merge pela coluna cidade

```python
# pegando os nomes das colunas
df_ndwi.columns

df_ndwi = df_ndwi[['cidade', 'ndwi']]

# conferindo
df_ndwi.columns
```

```python
df_merge = pd.merge(df_meteo, df_ndwi, on='cidade', how='left')
```

# conferindo e salvando

```python
print(f" Dataset final: {len(df_merge)} registros")
print(f" Colunas: {df_merge.columns.tolist()}")

# Checa se NDWI foi adicionado corretamente
print("\n=== NDWI POR CIDADE ===")
print(df_merge.groupby('cidade')['ndwi'].first())

# Checa nulos
print("\n=== VALORES NULOS ===")
print(df_merge.isnull().sum())
```

```python
df_merge.to_csv('../data/processado/dados_completos.csv', index=False)

print('salvo')
```
